Simple Encryption and Decryption:
Implement a basic Caesar cipher to encrypt and decrypt
messages

```python
import string
print("Welcome to cipher")
plaintext = input("Enter the text to be encrypted")
enc=''
dec=''
k = int(input("Enter the key"))
alphabet_dict = {}
def get_key(dict1,val):
  for k,v in dict1.items():
    if v == val:
      return k
  return None
for index, letter in enumerate(string.ascii_lowercase):
  alphabet_dict[letter] = index
for s in plaintext:
  i=(alphabet_dict[s]+k)%26
  enc+=get_key(alphabet_dict,i)
print(enc)
for s in enc:
  j=(alphabet_dict[s]-k)%26
  dec+=get_key(alphabet_dict,j)
print(dec)
```

```python
import random
print("digital dice")
score = 0
wins=0
losses =0
while True:
  ip=input("Enter 1 to continue, anything else to exit")
  if ip =='1':
    try:
      num = int(input("Guess the number"))
    except ValueError:
      print('invalid input')
      continue
    if num not in range(1,7):
      print("invalid input")
      continue
    dice = random.randint(1,6)
    if num == dice:
      print("Right Guess")
      wins+=1
      score+=1
    else:
      print("Wrong Guess")
      losses+=1
      score-=1
  else:
    print(f"wins:{wins}\nlosses:{losses}\ntotal score:{score}")
    break
```
