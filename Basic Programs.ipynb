## Shallow vs. Deep Copy

When you copy a complex object (like a list containing other lists), you can perform either a shallow copy or a deep copy. The difference lies in how nested objects are handled.

### Shallow Copy

A shallow copy creates a new compound object, but it doesn't create copies of the nested objects within the original object. Instead, it copies the references to the nested objects. This means that if you modify a nested object in the shallow copy, the original object will also be affected, and vice versa.

You can perform a shallow copy using methods like slicing (`[:]`), the `copy()` method of a list, or the `copy.copy()` function.

```python
import copy

original_list = [[1, 2, 3], [4, 5, 6]]
shallow_copy_list = original_list[:] # Using slicing for shallow copy

print("Original list:", original_list)
print("Shallow copy:", shallow_copy_list)

# Modify a nested object in the shallow copy
shallow_copy_list[0][0] = 99

print("\nAfter modifying shallow copy:")
print("Original list:", original_list)
print("Shallow copy:", shallow_copy_list)
```

As you can see, changing the nested list in the shallow copy also changed the nested list in the original list.

### Deep Copy

A deep copy creates a new compound object and recursively creates copies of all nested objects found in the original object. This means that a deep copy is completely independent of the original object, and changes to the deep copy will not affect the original object.

You can perform a deep copy using the `copy.deepcopy()` function.

```python
import copy

original_list_deep = [[1, 2, 3], [4, 5, 6]]
deep_copy_list = copy.deepcopy(original_list_deep) # Using deepcopy

print("Original list:", original_list_deep)
print("Deep copy:", deep_copy_list)

# Modify a nested object in the deep copy
deep_copy_list[0][0] = 99

print("\nAfter modifying deep copy:")
print("Original list:", original_list_deep)
print("Deep copy:", deep_copy_list)
```

In this case, modifying the nested list in the deep copy did not affect the original list.

In summary:
- **Shallow Copy:** Copies the top-level structure and references to nested objects. Changes to nested objects affect both copies.
- **Deep Copy:** Copies the entire structure, including all nested objects. Changes to the copy do not affect the original.

## Difference between `append()` and `extend()` for Lists

Both `append()` and `extend()` are list methods used to add elements to a list, but they do so in different ways:

### `append()`

The `append()` method adds a single element to the end of a list. If you append a list to another list using `append()`, the entire list you are appending will be added as a single element at the end.

```python
list1 = [1, 2, 3]
list2 = [4, 5]

list1.append(list2)
print(list1)
```

As you can see, `list2` was added as a single nested list within `list1`.

### `extend()`

The `extend()` method adds the elements of an iterable (like another list, tuple, or string) to the end of the current list. It effectively extends the list by adding each element of the iterable individually.

```python
list3 = [1, 2, 3]
list4 = [4, 5]

list3.extend(list4)
print(list3)

string_to_extend = "abc"
list3.extend(string_to_extend)
print(list3)
```

In summary:

- `append()` adds a single element (which can be a list or other iterable) to the end of the list.
- `extend()` adds all the individual elements of an iterable to the end of the list.

Note: single element of numeric type cannot be added using extend, it thows error
